```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, SpectralClustering
from mpl_toolkits.mplot3d import Axes3D

```

```python
# Define dataset X
X = np.array([
    [3, 1, 1, 4, 1.5, 0.12, 0, 0.03, 0.1],
    [1, 4, 1, 1, 3, 2, 6, 0.5, 4],
    [0.1, 0.02, 0, 0.1, 0, 1.9, 3.5, 1, 3]
]).T  # Transpose to get proper shape

```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='r', marker='x')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot of Dataset X')
plt.show()
```

```python
# Visualization of dataset in 3D using scatter3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='b', marker='o')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot of Dataset X')
plt.show()
```

```python
# (b) Apply Hierarchical Clustering (Ward's method)
hc = AgglomerativeClustering(n_clusters=2, linkage='ward')
hc_labels = hc.fit_predict(X)
```

```python
# Apply Spectral Clustering with adjusted n_neighbors to avoid ValueError
sc = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=min(5, len(X) - 1), random_state=42)
sc_labels = sc.fit_predict(X)
```

```python

# Plot results
fig, axes = plt.subplots(1, 2, figsize=(12, 5), subplot_kw={'projection': '3d'})
for ax, labels, title in zip(axes, [hc_labels, sc_labels], ["Hierarchical Clustering (Ward)", "Spectral Clustering"]):
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', marker='o')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title(title)
plt.show()
```

```python
# Plot results using scatter3D
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c=hc_labels, cmap='viridis', marker='o')
ax1.set_title("Hierarchical Clustering (Ward)")
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Z-axis')

ax2.scatter(X[:, 0], X[:, 1], X[:, 2], c=sc_labels, cmap='viridis', marker='o')
ax2.set_title("Spectral Clustering")
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_zlabel('Z-axis')

plt.show()
```
